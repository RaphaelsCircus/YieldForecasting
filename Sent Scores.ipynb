```python
# import modules
import os
import sys
import pandas as pd
import numpy as np
import transformers
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import pipeline
import re

pd.options.display.max_colwidth = 100
```

```python
PATH = os.getcwd()
DATANAME_1 = 'interest_data.csv'
DATANAME_2 = 'treasury_data.csv'
DATANAME_3 = 'yield_c_data.csv'

DATANAME_4 = 'labour_data.csv'
DATANAME_5 = 'unemployment_data.csv'

DATANAME_6 = 'inflation_data.csv'
```

```python
#load data
#merge these three
interest_data = pd.read_csv(PATH + '/' + DATANAME_1, low_memory=False)
treasury_data = pd.read_csv(PATH + '/' + DATANAME_2, low_memory=False)
yield_data = pd.read_csv(PATH + '/' + DATANAME_3, low_memory=False)

#merge these two
labour_data = pd.read_csv(PATH + '/' + DATANAME_4, low_memory=False)
unemployment_data = pd.read_csv(PATH + '/' + DATANAME_5, low_memory=False)

#standalone sentiment
inflation_data = pd.read_csv(PATH + '/' + DATANAME_5, low_memory=False) 
```

```python
 # Function to remove URLs from text
def remove_urls(text):
    url_pattern = re.compile(r'http\S+|www\.\S+|\b\w+\.(com|net|org|edu|gov|uk)\b', re.IGNORECASE)

    return url_pattern.sub(r'', text)
```

```python
# Apply the function to the 'tweet' column
interest_data['tweet'] = interest_data['tweet'].apply(remove_urls)
treasury_data['tweet'] = treasury_data['tweet'].apply(remove_urls)
yield_data['tweet'] = yield_data['tweet'].apply(remove_urls)

labour_data['tweet'] = labour_data['tweet'].apply(remove_urls)
unemployment_data['tweet'] = unemployment_data['tweet'].apply(remove_urls)

inflation_data['tweet'] = inflation_data['tweet'].apply(remove_urls)
```
